```python
data = pd.DataFrame({
    'red_cards_per_game' : raw_data['redCards'] / raw_data['games'],
    'skin_color'         : (raw_data['rater1'] + raw_data['rater2']) / 2,
    'height'             : raw_data['height'],
    'weight'             : raw_data['weight'],
    'age'                : (pd.to_datetime('2013') - raw_data['birthday']) / pd.Timedelta(days=365.25),
    'position'           : raw_data['position'],
    'meanIAT'            : raw_data['meanIAT'],
    'meanExp'            : raw_data['meanExp'],
    'club'               : raw_data['club'],
    'refCountry'         : raw_data['refCountry'],
})

data.dropna(inplace=True)

data
```

```python
sns.barplot(data=data, x='red_cards', y='skin_color', errwidth=1, capsize=0.2, n_boot=1000, ci=95)

plt.xlabel('Red cards per game')
plt.ylabel('Conversion Rate')
```

```python
from sklearn.linear_model import LinearRegression

data_encoded = data.copy(deep=True)
data_encoded = pd.get_dummies(data_encoded, sparse=True, dtype=bool)

target = data_encoded.columns[0]
predictors = data_encoded.columns[1:]

y = data_encoded[target].values
X = data_encoded[predictors].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)

reg = LinearRegression().fit(X_train, y_train)
```
